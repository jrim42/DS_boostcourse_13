### 데이터 분석 준비하기
[10 minutes to pandas](https://pandas.pydata.org/pandas-docs/version/1.0.0/getting_started/10min.html)

1. pandas
    - 수식 계산 및 데이터의 시각화가 가능한 데이터 분석도구
    - 대용량의 데이터 분석 가능
    - `import pandas as pd`를 사용하여 pandas 대신 pd로 사용 가능

2. methods
    - data frame: 2차원 테이블 형식의 데이터 구조 (matrix)
    - series: 1차원 배열과 유사한 데이터 구조
    - reshaping
        - `sort_values`: 특정 기준/조건으로 정렬해준다
        - `drop`: 특정 열/행을 제거해준다
    - group data
        - `groupby`: 데이터프레임에서 특정 열(의 조합)을 기준으로 그룹화해준다
        - `pivot_table`: 데이터프레임을 재구성하여 새로운 표 형태로 만들어준다

    - plotting: 데이터의 시각화

3. 경로 + 파일 불러오기
    - 절대경로와 상대경로
    - `pd.read_csv("파일이름", encoding="cp949")`로 파일을 불러올 수 있다
    - `encoding="cp949"`를 사용하면 한글이 깨지지 않는다
